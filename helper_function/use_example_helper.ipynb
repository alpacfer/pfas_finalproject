# Testing the library

```python
import bbhelp as bb
```

```python
# Paths
images_folder = r'C:\Users\User\Documents\dataset\pfas\34759_final_project_rect\seq_02\image_02\data'
labels_file = r'C:\Users\User\Documents\dataset\pfas\34759_final_project_rect\seq_02\labels.txt'
```

# Display images

```python
# Display the image
image_index = 50
bb.display_image(image_index, images_folder, labels_file) # Default display type is 'bbox'
bb.display_image(image_index, images_folder, labels_file, display_type='center') # Display type is 'center'
```

# Extract info from image

```python
# Image you want to extract information from
frame_info = bb.extract_image_info(image_index, labels_file)

# Track ID you want to access
desired_track_id = 5

# Access the center coordinates for track_id 5
center_coordinates = frame_info.get(desired_track_id, {}).get('center')

if center_coordinates:
    print(f"Center of track_id {desired_track_id}: {center_coordinates}")
else:
    print(f"No object with track_id {desired_track_id} found in this frame.")

# Access the bounding box for track_id 5
bounding_box = frame_info.get(desired_track_id, {}).get('bbox')

if bounding_box:
    print(f"Bounding box of track_id {desired_track_id}: {bounding_box}")
else:
    print(f"No object with track_id {desired_track_id} found in this frame.")

# Access the type of object for track_id 5
object_type = frame_info.get(desired_track_id, {}).get('type')

if object_type:
    print(f"Type of object with track_id {desired_track_id}: {object_type}")
else:
    print(f"No object with track_id {desired_track_id} found in this frame.")
    
```
